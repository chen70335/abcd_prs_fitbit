## calculate income-to-needs ratio

```python
%%R
## income band median calculation
income_medians <- c(
  `1` = 2500,            # Less than $5,000
  `2` = (5000 + 11999) / 2,  # $5,000 through $11,999
  `3` = (12000 + 15999) / 2, # $12,000 through $15,999
  `4` = (16000 + 24999) / 2, # $16,000 through $24,999
  `5` = (25000 + 34999) / 2, # $25,000 through $34,999
  `6` = (35000 + 49999) / 2, # $35,000 through $49,999
  `7` = (50000 + 74999) / 2, # $50,000 through $74,999
  `8` = (75000 + 99999) / 2, # $75,000 through $99,999
  `9` = (100000 + 199999) / 2, # $100,000 through $199,999
  `10` = 200000,          # $200,000 and greater (use $200,000 as the minimum)
  `999` = NA,             # Don't know
  `777` = NA              # Refuse to answer
)

# Create a vector of federal poverty income levels for family sizes up to 12 for 2024
poverty_levels_2024 <- c(
  `1` = 15060,                           # Individual
  `2` = 20440,                           # Family of 2
  `3` = 25820,                           # Family of 3
  `4` = 31200,                           # Family of 4
  `5` = 36580,                           # Family of 5
  `6` = 41960,                           # Family of 6
  `7` = 47340,                           # Family of 7
  `8` = 52720,                           # Family of 8
  `9` = 52720 + 5380,                    # Family of 9
  `10` = 52720 + 2 * 5380,               # Family of 10
  `11` = 52720 + 3 * 5380,               # Family of 11
  `12` = 52720 + 4 * 5380,                # Family of 12
  `13` = 52720 + 5 * 5380,               # Family of 13
  `14` = 52720 + 6 * 5380,                # Family of 14
  `15` = 52720 + 7 * 5380,               # Family of 15
  `16` = 52720 + 8 * 5380                # Family of 16
)


```

```python
%%R
# actually get income-to-needs ratio
df.abcd_p_demo <- df.abcd_p_demo %>%
  mutate(
    income = sapply(demo_comb_income_v2, function(income) income_medians[as.character(income)]),
    poverty_level = sapply(demo_roster_v2, function(size) poverty_levels_2024[as.character(size)]),
    income_to_needs_ratio = income / poverty_level
  )
```

## (Redacted) turn site ids into covariance matrix to fit as random effect in model

```python
%%R
# Optimized symmetric matrix computation
subjects <- df.abcd_site_ids$src_subject_id
sites <- df.abcd_site_ids$site_id_l
n <- length(subjects)

# Initialize the matrix with zeros
subject_site_matrix <- matrix(0, nrow = n, ncol = n, dimnames = list(subjects, subjects))

# Compute only the upper triangle
for (i in 1:(n - 1)) {
  for (j in (i + 1):n) {
    if (sites[i] == sites[j]) {
      subject_site_matrix[i, j] <- 1 # equals to 1 if in same site; not actua site number
    }
  }
}

# for (i in 1:n) {
#   subject_site_matrix[i, i] <- 1
# }

# Mirror the upper triangle to the lower triangle
subject_site_matrix <- subject_site_matrix + t(subject_site_matrix)


diag(subject_site_matrix) <- 1

# Optional: Convert to data frame for better viewing
subject_site_matrix_df <- as.data.frame(subject_site_matrix)
```

```python
# prompt: for all unique src_subject_ids in above cbcl df, if participant had a column dep_score below < 65 when eventname is at "2_year_follow_up" and dep_score >= 65 when eventname is at "3_year_follow_up", mark a new column "conversion" to 1. Otherwise mark it to 0

%%R
# create a new dataframe
src_subject_ids = c()
conversions = c()

# Create a new column 'conversion' and initialize it to 0


# Iterate through unique src_subject_ids
for (subject_id in unique(df.data_CBCL_no_subset$src_subject_id)) {
  # Subset data for the current subject
  subject_data <- df.data_CBCL_no_subset[df.data_CBCL_no_subset$src_subject_id == subject_id, ]

  # Check for the specified conditions
  year2_data <- subject_data[subject_data$eventname == "2_year_follow_up_y_arm_1", ]
  year3_data <- subject_data[subject_data$eventname == "3_year_follow_up_y_arm_1", ]

  # Check if both timepoints exist for the subject
  if (nrow(year2_data) > 0 && nrow(year3_data) > 0) {
    src_subject_ids <- c(src_subject_ids, subject_id)

    # Use any of the depression scores (choose one consistently)
    dep_score_year2 <- year2_data$cbcl_scr_syn_anxdep_t
    dep_score_year3 <- year3_data$cbcl_scr_syn_anxdep_t

    # Check for conversion condition
    if (!is.na(dep_score_year2) && !is.na(dep_score_year3) && dep_score_year2 < 65 && dep_score_year3 >= 65) {
      # Update the 'conversion' column for all rows of this subject
      conversions <- c(conversions, 1)
    } else {
      conversions <- c(conversions, 0)
    }
  }
}


```

### use CBCL as continuous outcome

```python
%%R
# function to subset df by eventname easily by specifying year as 2, 3, or 4
filter_df_by_year <- function(df, year) {
  eventname_mapping <- list(
    "2" = "2_year_follow_up_y_arm_1",
    "3" = "3_year_follow_up_y_arm_1",
    "4" = "4_year_follow_up_y_arm_1"
  )

  selected_eventname <- eventname_mapping[[year]]

  df %>%
    filter(eventname == selected_eventname)
}
```

```python
%%R
df.data_CBCL_cont_year_2 = filter_df_by_year(df.data_CBCL_no_subset, "2") %>%
  select(-eventname)
print(colnames(df.data_CBCL_cont_year_2))
print(nrow(df.data_CBCL_cont_year_2))
```

### z-score PRS scores

```python
%%R
# write a function to z score a prs file by ancestry
z_score_within_ancestry <- function(prs_df, ancestry_df, prs_colname) {
  # Merge the data frames on 'subjectid' to align prs scores with ancestry groups
  merged_df <- merge(prs_df, ancestry_df, by = "src_subject_id")

  # Z-score the prs_score within each AncestryGroup
  output_colname = paste0("z_scored_", prs_colname)

  merged_df <- merged_df %>%
    group_by(AncestryGroup) %>%
    mutate(!!output_colname := scale(.data[[prs_colname]], center = TRUE, scale = TRUE)) %>%
    ungroup() %>%
    select(-.data[[prs_colname]], -"AncestryGroup")

  # Return the result with selected columns
  return(merged_df[, ])
}

# Example usage:
# Assuming `prs_df` is the data frame with 'subjectid' and 'prs_score'
# and `ancestry_df` is the data frame with 'subjectid' and 'AncestryGroup'
df.prscsx_meta_z_prs <- z_score_within_ancestry(df.prscsx_meta_prs, df.data_ancestry, "prs")
df.prscsx_eur_z_prs <- z_score_within_ancestry(df.prscsx_eur_prs, df.data_ancestry, "prs")
df.prscsx_afr_z_prs <- z_score_within_ancestry(df.prscsx_afr_prs, df.data_ancestry, "prs")
df.prscsx_amr_z_prs <- z_score_within_ancestry(df.prscsx_amr_prs, df.data_ancestry, "prs")

# print(result)
df.prscsx_meta_z_prs_11_25 <- z_score_within_ancestry(df.prscsx_meta_prs_11_25, df.data_ancestry, "prs")
df.prscsx_eur_z_prs_11_25 <- z_score_within_ancestry(df.prscsx_eur_prs_11_25, df.data_ancestry, "prs")

df.prscsx_meta_z_prs_12_02 <- z_score_within_ancestry(df.prscsx_meta_prs_12_02, df.data_ancestry, "prs")
df.prscsx_eur_z_prs_12_02 <- z_score_within_ancestry(df.prscsx_eur_prs_12_02, df.data_ancestry, "prs")
df.prscsx_afr_z_prs_12_02 <- z_score_within_ancestry(df.prscsx_afr_prs_12_02, df.data_ancestry, "prs")
df.prscsx_amr_z_prs_12_02 <- z_score_within_ancestry(df.prscsx_amr_prs_12_02, df.data_ancestry, "prs")

```

```python
%%R
# format datetime columns
  # select(-fit_ss_first_inbed_minutes, -fit_ss_outbed_minutes,
  #        -fit_ss_first_sleep_minutes, -fit_ss_wakeup_minutes)
df.fitbit_w_slp_year2_nocvd_time_transformed = df.fitbit_w_slp_year2_nocvd %>%
  mutate(
      # - 12 hours because all fitbit data in ABCD were offset by 12 hours
      fit_ss_avg_first_inbed_hour_adjusted = hm(fit_ss_avg_first_inbed_hour),
      fit_ss_avg_outbed_hour_adjusted = hm(fit_ss_avg_outbed_hour),
      fit_ss_avg_first_sleep_hour_adjusted = hm(fit_ss_avg_first_sleep_hour),
      fit_ss_avg_wakeup_hour_adjusted = hm(fit_ss_avg_wakeup_hour),
      # fit_ss_avg_first_inbed_hour_adjusted = fit_ss_avg_first_inbed_hour - hours(12),
      # fit_ss_avg_outbed_hour_adjusted = fit_ss_avg_outbed_hour - hours(12),
      # fit_ss_avg_first_sleep_hour_adjusted = fit_ss_avg_first_sleep_hour - hours(12),
      # fit_ss_avg_wakeup_hour_adjusted = fit_ss_avg_wakeup_hour - hours(12),
  ) %>%
  # generate total time in bed/ asleep features
  # add sleep latency and sleep inertia features
  mutate(
      fit_ss_sleep_period_minutes = fit_ss_sleep_period_minutes / fit_ss_day_count,
      # fit_ss_total_time_inbed = as.numeric(if_else(
      #     fit_ss_avg_outbed_hour_adjusted >= fit_ss_avg_first_inbed_hour_adjusted,
      #     fit_ss_avg_outbed_hour_adjusted - fit_ss_avg_first_inbed_hour_adjusted,
      #     fit_ss_avg_outbed_hour_adjusted + hours(24) - fit_ss_avg_first_inbed_hour_adjusted
      # ), "mins"),
      # fit_ss_total_time_asleep = as.numeric(if_else(
      #     fit_ss_avg_wakeup_hour_adjusted >= fit_ss_avg_first_sleep_hour_adjusted,
      #     fit_ss_avg_wakeup_hour_adjusted - fit_ss_avg_first_sleep_hour_adjusted,
      #     fit_ss_avg_wakeup_hour_adjusted + hours(24) - fit_ss_avg_first_sleep_hour_adjusted
      # ), "mins"),
      # fit_ss_sleep_latency = as.numeric(if_else(
      #     fit_ss_avg_first_sleep_hour_adjusted >= fit_ss_avg_first_inbed_hour_adjusted,
      #     fit_ss_avg_first_sleep_hour_adjusted - fit_ss_avg_first_inbed_hour_adjusted,
      #     fit_ss_avg_first_sleep_hour_adjusted + hours(24) - fit_ss_avg_first_inbed_hour_adjusted
      # ), "mins"),
      # fit_ss_sleep_inertia = as.numeric(if_else(
      #     fit_ss_avg_outbed_hour_adjusted >= fit_ss_avg_wakeup_hour_adjusted,
      #     fit_ss_avg_outbed_hour_adjusted - fit_ss_avg_wakeup_hour_adjusted,
      #     fit_ss_avg_outbed_hour_adjusted + hours(24) - fit_ss_avg_wakeup_hour_adjusted
      # ), "mins"),
  ) %>%
  # generate hour of day features (for start and end inbed/ asleep)
  mutate(
      fit_ss_first_asleep_hour_adjusted = if_else(
          12 <= hour(fit_ss_avg_first_sleep_hour_adjusted),
          hour(fit_ss_avg_first_sleep_hour_adjusted - hours(12)),
          hour(fit_ss_avg_first_sleep_hour_adjusted + hours(12))
      ),
      fit_ss_end_asleep_hour_adjusted = if_else(
          12 <= hour(fit_ss_avg_wakeup_hour_adjusted),
          hour(fit_ss_avg_wakeup_hour_adjusted - hours(12)),
          hour(fit_ss_avg_wakeup_hour_adjusted + hours(12))
      )
  )



```

```python
%%R
# input: explanation
# data_df: this function takes in a dataframe for abcd subjects that contain all information
# then, reorders the dataframe to match the subset genetic related matrices
# grm_df: Also, takes in a GRM dataframe, then outputs a subset that matches the row ids of the
# previous input abcd subject dataframe

subset_and_reorder_grm <- function(data_df, grm_df, id_column="IID_withPrefix") {
  row_indices <- match(data_df[[id_column]], rownames(grm_df))
  grm_subset <- grm_df[row_indices, row_indices]
  reordered_data_df <- data_df[match(colnames(grm_subset), data_df[[id_column]]),]
  rownames(reordered_data_df) <- reordered_data_df[[id_column]]
  return(list(reordered_data_df = reordered_data_df, grm_subset = grm_subset))
}

# Example usage:
# result <- subset_and_reorder_grm(df.prscsx_meta_z_prs, df.abcd_all_grms)
# The reordered data frame and subsetted GRM can be accessed with:
# result$reordered_data_df
# result$grm_subset
```

```python
%%R
# function to combine all variables and run GENESIS mixed effects model
# can handle interactions between two variables

# function inputs
# main_df: Dataframe containing the main effect variable dataset with individual-level data
# main_colname: The column name in main_df representing the main predictor variable (e.g., genetic score)
# mdd_df: Dataframe containing MDD diagnosis data
# mdd_colname: The column name in mdd_df representing MDD diagnosis (binary or continuous)
# binary: Logical indicating if the outcome variable is binary (TRUE for logistic regression, FALSE for linear regression)
# age_year: The age (in years) used to filter data (this is the threshold year)
# ancestry: Subset of ancestry to filter by (default is "all", other options could be specific groups)
# covar_grm: Logical indicating if GRM (Genetic Relationship Matrix) should be included as a covariate in the model
# interactions: Logical indicating if interaction between main predictor and another variable should be included
# interact_df: Dataframe that contains the interacting column (not used in this function, may need clarification)
# interact_colname: Name of the column to interact with the main_colname (only used if interactions = TRUE)
# verbose: Logical flag to print additional information for debugging
combine_variables_and_run_lm = function(main_df, main_colname,
                                     mdd_df, mdd_colname, binary, age_year,
                                     ancestry="all", covar_grm=TRUE,
                                     interactions=FALSE, interact_df=NULL, interact_colname=NULL,
                                     verbose=FALSE) {



  # get interview age at year 3
  # TODO: make year an input variable
  df.abcd_y_lt_year_3 = filter_df_by_year(df.abcd_y_lt_no_subset, age_year) %>%
    select(-eventname)

  dfs_to_merge <- list(main_df, mdd_df, df.abcd_p_demo, df.abcd_y_lt_year_3,
                     df.gish_p_gi, df.data_ancestry_PC)

  # Add interact_df if interactions is TRUE
  if (interactions) {
    dfs_to_merge <- append(dfs_to_merge, list(interact_df))
  }
  # Perform the merge using Reduce
  df.merged_abcd_z_prs_demo_ancestry_pc_mdd_lifetime <- Reduce(function(x, y) merge(x, y, by = "src_subject_id", all = FALSE),
                                                                dfs_to_merge)

  if (ancestry != "all") {
    df.merged_abcd_z_prs_demo_ancestry_pc_mdd_lifetime = merge(df.merged_abcd_z_prs_demo_ancestry_pc_mdd_lifetime,
                                                            df.data_ancestry, by = "src_subject_id") %>%
                                                              filter(AncestryGroup == ancestry) %>%
                                                              select(-AncestryGroup)
  }
  if (verbose) {
    print("check the columns of the combined dataframe")
    print(colnames(df.merged_abcd_z_prs_demo_ancestry_pc_mdd_lifetime))
    print("check mdd dx analysis sample size and lifetime mdd dx spread")
    print(table(df.merged_abcd_z_prs_demo_ancestry_pc_mdd_lifetime[[mdd_colname]]))
  }

  reorder_with_grm_result <- subset_and_reorder_grm(df.merged_abcd_z_prs_demo_ancestry_pc_mdd_lifetime, df.abcd_all_grms)
  if (verbose) {
    print("check reordering worked below:")
    print("first, check reordered_data_df:")
    print(nrow(reorder_with_grm_result$reordered_data_df))
    print(rownames(reorder_with_grm_result$reordered_data_df)[0:10])

    print("second, check if reordered grm subset has same order of IIDs")
    print(nrow(reorder_with_grm_result$grm_subset))
    print(rownames(reorder_with_grm_result$grm_subset)[0:10])
  }


  if (interactions) {
      reorder_with_grm_result$reordered_data_df = reorder_with_grm_result$reordered_data_df %>%
        mutate(main_interact = .data[[main_colname]] * .data[[interact_colname]])
      selected_covars = c(main_colname, interact_colname, "main_interact", "interview_age", "income_to_needs_ratio",
                                              "demo_gender_id_v2", paste0("V", 1:8))
  } else {
      selected_covars = c(main_colname, "interview_age", "income_to_needs_ratio",
                                              "demo_gender_id_v2", paste0("V", 1:8))
  }

  model = fitNullModel(reorder_with_grm_result$reordered_data_df,
                                   outcome = mdd_colname,
                                   covars = selected_covars,
                                   cov.mat = if (covar_grm) reorder_with_grm_result$grm_subset else NULL,
                                   # cov.mat = reorder_with_site_result$grm_subset,
                                   verbose = verbose,
                                   family = if (binary) "binomial" else "gaussian") # TODO: add option to use binary or continous outcome
  return(model)
}




```

## Function to create table output of each main variable's effect size, logOR, and p-values in the mixed effects model

```python
%%R
generate_model_results_table = function(df, mdd_df, mdd_colname, interaction) {
    # to output to a dataframe
    cols_final = list()
    beta_list = list()
    p_val_list = list()


    # these columns will each be a fixed effect in separate GENESIS models
    cols_of_interest = df %>%
      select(where(is.numeric)) %>%
      colnames()
    for (col in cols_of_interest) {
        df_col_only = df %>%
          select(src_subject_id, IID_withPrefix, .data[[col]])
        # combine_variables_and_run_lm = function(main_df, main_colname,
                                    #  mdd_df, mdd_colname, binary, age_year,
                                    #  ancestry="all", covar_grm=TRUE,
                                    #  interactions=FALSE, interact_df=NULL, interact_colname=NULL,
                                    #  verbose=FALSE) {
        df.prscs_z_prs_noIID = df.prscs_z_prs %>%
          select(-IID_withPrefix)
        if (interaction) {
          model = combine_variables_and_run_lm(df_col_only, col,
                                              mdd_df, mdd_colname, binary=FALSE, age_year=2,
                                              ancestry="all", covar_grm=TRUE,
                                              interactions=TRUE, interact_df=df.prscs_z_prs_noIID, interact_colname="z_scored_prs_phi_auto",
                                              verbose=FALSE)
        } else {
          model = combine_variables_and_run_lm(df_col_only, col,
                                            mdd_df, mdd_colname, binary=FALSE, age_year=2,
                                            ancestry="all", covar_grm=TRUE)
        }


        # model = combine_variables_and_run_lm(df_col_only, col,
        #                               mdd_df, mdd_colname, binary=FALSE, age_year=2,
        #                               ancestry="all", covar_grm=TRUE,
        #                               interactions=FALSE)
        results = model$fixef
        cols_final = c(cols_final, col)
        beta = results[2, 1]
        beta_list = c(beta_list, beta)
        p_val = results[2, 4]
        p_val_list = c(p_val_list, p_val)

        if (interaction) {
          cols_final = c(cols_final, paste0("prs", col))
          beta = results[3, 1]
          beta_list = c(beta_list, beta)
          p_val = results[3, 4]
          p_val_list = c(p_val_list, p_val)

          cols_final = c(cols_final, paste0("prs_interact", col))
          beta = results[4, 1]
          beta_list = c(beta_list, beta)
          p_val = results[4, 4]
          p_val_list = c(p_val_list, p_val)
        }
    }

    df.output = data.frame(
        main_variable = I(cols_final),
        beta = I(beta_list),
        p_val = I(p_val_list)
    )

    # df.output = df.output %>%
    #   arrange(desc(beta))

    return(df.output)
}
```

```python
%%R
# function to subset df by eventname easily by specifying year as 2, 3, or 4
filter_df_by_year <- function(df, year) {
  eventname_mapping <- list(
    "2" = "2_year_follow_up_y_arm_1",
    "3" = "3_year_follow_up_y_arm_1",
    "4" = "4_year_follow_up_y_arm_1"
  )

  selected_eventname <- eventname_mapping[[year]]

  df %>%
    filter(eventname == selected_eventname)
}
```

```python
%%R
# check whether PRS or Fitbit is still significant even with interaction
process_interaction_results <- function(df) {
  independent_significances = c()
  n <- nrow(df)
  for (i in seq(1, n, by = 3)) {
    # Select the group of 3 rows (or less if at the end)
    subset_df <- df[i:min(i+2, n), ]
    interact_effect <- subset_df[1, ]
    prs_effect <- subset_df[2, ]
    fitbit_effect <- subset_df[3, ]

    if (interact_effect$p_val < 0.05) {
        if (prs_effect$p_val < 0.05 & fitbit_effect$p_val < 0.05) {
          independent_significances = c(independent_significances, c("both", "both", "both"))
        } else if (prs_effect$p_val < 0.05) {
          independent_significances = c(independent_significances, c("prs", "prs", "prs"))
        } else {
          independent_significances = c(independent_significances, c("fitbit", "fitbit", "fitbit"))
        }
    } else {
      independent_significances = c(independent_significances, c("none", "none", "none"))
    }
  }
  df$independent_significance = independent_significances
  return (df)
}
```

```python
%%R
df.fitbit_slp_sumstats_interact_results_year2 = process_interaction_results(df.fitbit_slp_sumstats_interact_results_year2)
df.fitbit_act_sumstats_interact_results_year2 = process_interaction_results(df.fitbit_act_sumstats_interact_results_year2)
```

## Visualization of beta scores for Fitbit in own model

```python
%%R
plot_main_effects_results <- function(df, direction, top_n) {
  # Filter rows with p-value < 0.05
  df$beta <- as.numeric(as.character(df$beta))
  if (direction == "positive") {
    df_filtered <- df %>%
      filter(p_val < 0.05) %>%
      arrange(desc(beta)) %>%
      head(top_n)  # Select top top_n
  } else {
    df_filtered <- df %>%
      filter(p_val < 0.05) %>%
      arrange(beta) %>%
      head(top_n)  # Select top top_n
  }

  # Create the plot
  ggplot(df_filtered, aes(x = reorder(main_variable, abs(beta)), y = beta, fill = independent_significance)) +
    geom_bar(stat = "identity") +
    coord_flip() +  # Flip coordinates to make the x-axis labels readable
    theme_minimal() +
    labs(x = "Main Variable", y = "Beta Value", title = "Top Interaction Effects (p < 0.05)") +
    theme(axis.text.x = element_text(angle = 90, hjust = 1)) + # Rotate x-axis labels vertically
    scale_fill_manual(values = c("prs" = "steelblue", "fitbit" = "darkorange",
                                 "both" = "forestgreen", "none" = "black"))  # Custom color scale
}
```

```python
%%R
plot_main_effects_results(df.fitbit_slp_sumstats_results_year2, "positive", 30)
```

## Visualization of Fitbit x PRS interactions

```python
%%R
plot_interaction_results <- function(df, direction, top_n) {
  df = df %>%
    filter(grepl("^prs_interact", main_variable))
  plot_main_effects_results(df, direction, top_n)
}
```
